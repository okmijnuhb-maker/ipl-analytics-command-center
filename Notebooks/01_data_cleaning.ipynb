```python
import pandas as pd
import numpy as np
import os
matches    = pd.read_csv(r'C:\my project\ipl-analytics-command-center\data\matches.csv')
deliveries = pd.read_csv(r'C:\my project\ipl-analytics-command-center\data\deliveries.csv')
print("Matches shape    :", matches.shape)
print("Deliveries shape :", deliveries.shape)
```

```python
# Check matches columns and first few rows
print("=== MATCHES COLUMNS ===")
print(matches.columns.tolist())
print("\n=== FIRST 3 ROWS ===")
print(matches.head(3))
```

```python
# Check deliveries columns and first few rows
print("=== DELIVERIES COLUMNS ===")
print(deliveries.columns.tolist())
print("\n=== FIRST 3 ROWS ===")
print(deliveries.head(3))
```

```python
print("=== MATCHES - Missing Values ===")
print(matches.isnull().sum())
print("\n=== DELIVERIES - Missing Values ===")
print(deliveries.isnull().sum())
```

```python
print("=== SEASONS ===")
print(sorted(matches['season'].unique()))
print("\n=== ALL TEAM NAMES ===")
teams = pd.unique(matches[['team1','team2']].values.ravel())
print(sorted(teams))
```

```python
#standardized team names
team_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
}

cols_m = ['team1','team2','winner','toss_winner']
cols_d = ['batting_team','bowling_team']

matches[cols_m]    = matches[cols_m].replace(team_map)
deliveries[cols_d] = deliveries[cols_d].replace(team_map)

print(sorted(pd.unique(matches[['team1','team2']].values.ravel())))
```

```python
#  Fix Date & Season
matches['date']   = pd.to_datetime(matches['date'])
matches['season'] = matches['season'].str[:4].astype(int)
print(matches[['date','season']].head())
print("\nSeasons:", sorted(matches['season'].unique()))
```

```python
# Clean Missing Values
matches['winner']          = matches['winner'].fillna('No Result')
matches['player_of_match'] = matches['player_of_match'].fillna('N/A')
matches['city']            = matches['city'].fillna(matches['venue'].str.split(',').str[0])

deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna('not_out')
deliveries['dismissal_kind']   = deliveries['dismissal_kind'].fillna('not_out')

print("Missing values after fix:")
print(matches[['winner','player_of_match','city']].isnull().sum())
print(deliveries[['player_dismissed','dismissal_kind']].isnull().sum())
```

```python
#  Remove Underscores from Columns & Values
matches.columns    = matches.columns.str.replace('_', ' ').str.title()
deliveries.columns = deliveries.columns.str.replace('_', ' ').str.title()

matches    = matches.apply(lambda col: col.str.replace('_', ' ').str.title() if col.dtype == 'object' else col)
deliveries = deliveries.apply(lambda col: col.str.replace('_', ' ').str.title() if col.dtype == 'object' else col)

print("Matches columns   :", matches.columns.tolist())
print("\nDeliveries columns:", deliveries.columns.tolist())
```

```python
#  Quick Check — Row Values
print(deliveries['Dismissal Kind'].unique())
print(deliveries['Toss Decision'].unique() if 'Toss Decision' in deliveries.columns else matches['Toss Decision'].unique())
```

```python
# 🔧Add Useful Columns
matches['Toss Win Match Win'] = (matches['Toss Winner'] == matches['Winner']).astype(int)
matches['Bat First Win']      = (matches['Result'] == 'Runs').astype(int)

for runs, name in [(4,'Four'),(6,'Six')]:
    deliveries[f'Is {name}'] = (deliveries['Batsman Runs'] == runs).astype(int)
deliveries['Is Boundary'] = deliveries['Is Four'] | deliveries['Is Six']

print(matches[['Toss Win Match Win','Bat First Win']].value_counts())
print(deliveries[['Is Boundary','Is Four','Is Six']].sum())
```

```python
#  Check & Remove Duplicates
for name, df in [('Matches', matches), ('Deliveries', deliveries)]:
    print(f"{name} duplicates: {df.duplicated().sum()}")
    df.drop_duplicates(inplace=True)
```

```python
#  Fix Data Types
matches[['Season','Result Margin','Target Runs']] = matches[['Season','Result Margin','Target Runs']].apply(pd.to_numeric, errors='coerce')
deliveries[['Is Wicket','Is Four','Is Six','Is Boundary']] = deliveries[['Is Wicket','Is Four','Is Six','Is Boundary']].astype('int8')

print(matches.dtypes)
```

```python
# Fix Over Column
deliveries['Over'] = deliveries['Over'] + 1
print(deliveries['Over'].unique())
```

```python
#  Drop Useless Columns
matches.drop(columns=['Umpire1','Umpire2','Target Overs','Method'], inplace=True)
print(matches.columns.tolist())
```

```python
#  Add IPL Season Winners
ipl_winners = {
    2007:'Rajasthan Royals', 2009:'Deccan Chargers', 2010:'Chennai Super Kings',
    2011:'Chennai Super Kings', 2012:'Kolkata Knight Riders', 2013:'Mumbai Indians',
    2014:'Kolkata Knight Riders', 2015:'Mumbai Indians', 2016:'Sunrisers Hyderabad',
    2017:'Mumbai Indians', 2018:'Chennai Super Kings', 2019:'Mumbai Indians',
    2020:'Mumbai Indians', 2021:'Chennai Super Kings', 2022:'Gujarat Titans',
    2023:'Chennai Super Kings', 2024:'Kolkata Knight Riders'
}

matches['Ipl Winner'] = matches['Season'].map(ipl_winners)
print(matches[['Season','Ipl Winner']].drop_duplicates().sort_values('Season'))
```

```python
# Save Cleaned Data
path = r'C:\my project\ipl-analytics-command-center\data'
matches.to_csv(f'{path}\matches_clean.csv', index=False)
deliveries.to_csv(f'{path}\deliveries_clean.csv', index=False)

print(f"Matches saved    : {matches.shape}")
print(f"Deliveries saved : {deliveries.shape}")
```
